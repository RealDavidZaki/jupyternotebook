```python
from tabulate import tabulate
import pandas as pd
from io import StringIO

# Function to display table with title and source
def display_table_with_title(title, source, text_data):
    print(f"\n\x1b[1m{title}\x1b[0m")  # Bold formatting for the title
    print(source)  # Display source information
    df = pd.read_csv(StringIO(text_data), sep=',')
    print(tabulate(df, headers='keys', tablefmt='pretty'))
```

According to Ellah, Achor, and Enemarie (2019), the study conducted in Benue State utilized various methods to measure memory strength and attention lapses.

A correlational survey design was employed within Otukpo, Ogbadibo, and Okpokwu Local Government Areas in Benue State. The study targeted a population of 2138 science students in public senior secondary schools within the specified areas. The sample consisted of 450 SSI students drawn from 24 public senior secondary schools, selected using a simple random sampling technique facilitated by a lucky dip with replacement method.

Twenty-four intact classes from the secondary schools in the specified Local Government Areas were purposively selected to meet the research objectives. The research utilized the Wechsler Adult Intelligence Scale (WAIS) backward digit span task and the Attention Lapses Clicker (ALC) for data collection. The WAIS backward digit span task assessed working memory by presenting a series of numbers and letters on a computer screen. Respondents were tasked with orally recalling the numbers in ascending order and then the letters in alphabetical order. Before the formal assessment, participants underwent six short practice trials. The series' complexity increased incrementally based on correct responses within a set of three, concluding after three unsuccessful attempts at a given set.

Utilizing the Attention Lapses Clicker (ALC), students reported attention lapses by pressing buttons corresponding to different lapse durations: one for lapses lasting 1 minute or less, another for lapses between 2 to 3 minutes, and a third for lapses exceeding 3 or more minutes. The recorded clicker-responses were compiled on a timeline to measure the average length of reported attention lapses.

Additionally, Students' Cognitive Ability Levels (SCAL) were deduced from students' SSI biology, chemistry, and physics promotion examination results. Baseline data collection involved converting raw scores into Z-scores and further converting them into T-scores to categorize students into low, average, and high ability levels. The instruments underwent face validation by three experts in Measurement and Evaluation from the Science Education Department at the University of Nigeria, Nsukka.
